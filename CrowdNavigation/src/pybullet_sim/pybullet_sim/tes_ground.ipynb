```python
import pickle
import pandas as pd

# Load trajectory
with open("expert_trajectories/expert_003.pkl", "rb") as f:
    traj = pickle.load(f)

print(type(traj))        # should be TrajectoryWithRew
print(dir(traj))         # see all attributes
```

```python
# Extract basic info
obs = traj.obs           # ndarray of observations
acts = traj.acts         # ndarray of actions
infos = traj.infos       # list of dicts
rews = traj.rews         # usually zero here

# Combine everything into a DataFrame
df = pd.DataFrame(infos)
df["action_v"] = acts[:, 0]
df["action_w"] = acts[:, 1]

# Optionally add flattened observations
obs_flat = obs[:-1].reshape((len(acts), -1))
obs_df = pd.DataFrame(obs_flat, columns=[f"obs_{i}" for i in range(obs_flat.shape[1])])
df = pd.concat([df, obs_df], axis=1)

df.head()
```
